```python
print("="*40)
print("SELAMAT DATANG DI Modern Device Studio")
print("="*40)
print("Aksesoris perlengkapan gadget dan laptop:")
print("1. Mouse Premium         - Rp 1.250.000")
print("2. Flashdisk 1 TB        - Rp 1.000.000")
print("3. Screen Cleaner        - Rp 50.000")
print("4. Powerbank 10000 mAh   - Rp 300.000")
print("5. Powerbank 20000 mAh   - Rp 500.000")
print("6. Wired Headset         - Rp 25.000")
print("7. Wireless Headset      - Rp 105.000")
print("8. Softcase              - Rp 55.000")
print("9. Charger Original      - Rp 150.000")
print("10. Keyboard             - Rp 350.000")
print("11. Antigores            - Rp 75.000")
print("-"*40)

# --------------------------
# FUNGSI PILIHAN PESANAN
# --------------------------
def pilih_barang(pilihan):
    if pilihan == 1:
        return "Mouse Premium", 1250000
    elif pilihan == 2:
        return "Flashdisk 1 TB", 1000000
    elif pilihan == 3:
        return "Screen Cleaner", 50000
    elif pilihan == 4:
        return "Powerbank 10000 mAh", 300000
    elif pilihan == 5:
        return "Powerbank 20000 mAh", 500000
    elif pilihan == 6:
        return "Wired Headset", 25000
    elif pilihan == 7:
        return "Wireless Headset", 105000
    elif pilihan == 8:
        return "Softcase", 55000
    elif pilihan == 9:
        return "Charger Original", 150000
    elif pilihan == 10:
        return "Keyboard", 350000
    elif pilihan == 11:
        return "Antigores", 75000
    else:
        return None, 0

# ========================
# PESANAN 1
# ========================
pilihan1 = int(input("Masukkan nomor pesanan pertama (1-11, isi 0 jika tidak): "))
nama1, harga1 = pilih_barang(pilihan1)
jumlah1 = 0
if nama1:
    jumlah1 = int(input(f"Berapa banyak {nama1} yang ingin dibeli? "))
subtotal1 = harga1 * jumlah1

# ========================
# PESANAN 2
# ========================
pilihan2 = int(input("Masukkan nomor pesanan kedua (1-11, isi 0 jika tidak): "))
nama2, harga2 = pilih_barang(pilihan2)
jumlah2 = 0
if nama2:
    jumlah2 = int(input(f"Berapa banyak {nama2} yang ingin dibeli? "))
subtotal2 = harga2 * jumlah2

# ========================
# PESANAN 3
# ========================
pilihan3 = int(input("Masukkan nomor pesanan ketiga (1-11, isi 0 jika tidak): "))
nama3, harga3 = pilih_barang(pilihan3)
jumlah3 = 0
if nama3:
    jumlah3 = int(input(f"Berapa banyak {nama3} yang ingin dibeli? "))
subtotal3 = harga3 * jumlah3


# ========================
# HITUNG TOTAL
# ========================
total_harga = subtotal1 + subtotal2 + subtotal3

print("\nDaftar Pesanan Anda:")
if nama1:
    print(f"- {jumlah1}x {nama1} = Rp {subtotal1:,}".replace(",", "."))
if nama2:
    print(f"- {jumlah2}x {nama2} = Rp {subtotal2:,}".replace(",", "."))
if nama3:
    print(f"- {jumlah3}x {nama3} = Rp {subtotal3:,}".replace(",", "."))

print(f"\nTotal harga sebelum diskon: Rp {total_harga:,}".replace(",", "."))

# ========================
# DISKON Shopping
# ========================
print("\nPilih kategori diskon:")
print("1. Diskon 70%")
print("2. Diskon 50%")
print("3. Diskon 30%")
kategori = int(input("Masukkan kategori diskon (1-3): "))

diskon = 0
if kategori == 1:
    diskon = 0.7
elif kategori == 2:
    diskon = 0.5
elif kategori == 3:
    diskon = 0.3
else:
    print("Kategori tidak valid, tidak ada diskon.")

potongan = total_harga * diskon
total_bayar = total_harga - potongan

print(f"\nHarga sebelum diskon : Rp {total_harga:,}".replace(",", "."))
print(f"Potongan Diskon        : Rp {potongan:,}".replace(",", "."))
print(f"Total Bayar            : Rp {total_bayar:,}".replace(",", "."))

# ========================
# PEMBAYARAN
# ========================
uang = int(input("\nMasukkan jumlah uang Anda: Rp "))
if uang >= total_bayar:
    kembalian = uang - total_bayar
    print("\n=== Struk Pembayaran ===")
    if nama1:
        print(f"{jumlah1}x {nama1} = Rp {subtotal1:,}".replace(",", "."))
    if nama2:
        print(f"{jumlah2}x {nama2} = Rp {subtotal2:,}".replace(",", "."))
    if nama3:
        print(f"{jumlah3}x {nama3} = Rp {subtotal3:,}".replace(",", "."))
    print(f"\nTotal Harga     : Rp {total_harga:,}".replace(",", "."))
    print(f"Diskon          : Rp {potongan:,}".replace(",", "."))
    print(f"Total Bayar     : Rp {total_bayar:,}".replace(",", "."))
    print(f"Uang Dibayar    : Rp {uang:,}".replace(",", "."))
    print(f"Kembalian       : Rp {kembalian:,}".replace(",", "."))
    print("Terima kasih telah berbelanja!")
else:
    kurang = total_bayar - uang
    print(f"Maaf, uang Anda kurang sebesar Rp {kurang:,}".replace(",", "."))

print("="*40)
```
