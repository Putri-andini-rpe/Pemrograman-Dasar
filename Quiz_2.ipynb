```python
# menampilkan deret bilangan ganajil dari 1 sampai 200 menggunakan while

bilangan_ganjil = 1
while bilangan_ganjil <= 200:
    print(bilangan_ganjil)
    bilangan_ganjil +=2

```

```python
# Progam menampilkan tabel perkalian menggunakan for
angka = int(input("Masukkan angka: "))
print("Tabel perkalian untuk", angka)
for i in range(1, 11):   # dari 1 sampai 10
    print(angka, "x", i, "=", angka * i)

# Program menampilkan tabel perkalian menggunakan while
angka = int(input("Masukkan angka; "))
print("Tabel perkalian untuk", angka)
i = 1
while i <= 10:
    print(angka, "x", i, "=", angka * i)
    i += 1
```
