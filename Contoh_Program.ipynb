```python
def sistem_penilaian_semua(data_siswa):
    # --- Tipe data non-primitif ---
    mata_pelajaran = ("Matematika", "Fisika", "Informatika")  # tuple
    tips_belajar = {
        "Matematika": "Latihan soal setiap hari",
        "Fisika": "Pahami konsep, jangan hafalan",
        "Informatika": "Perbanyak praktik coding"
    }  # dictionary
    tambahan = {"Jangan menyerah", "Terus semangat"}  # set

    # Perulangan untuk setiap siswa (data_siswa = list of dict)
    for siswa in data_siswa:
        nama = siswa["nama"]
        nilai = siswa["nilai"]
        status_lulus = False
        pesan = ""
        rekomendasi = []

        # --- Percabangan IF, ELIF, ELSE ---
        if nilai >= 90:
            pesan = "Nilai kamu sangat baik! üèÜ"
            status_lulus = True
        elif nilai >= 70:
            pesan = "Kamu lulus, tapi masih bisa ditingkatkan."
            status_lulus = True
        else:
            pesan = "Kamu belum lulus, tetap semangat belajar!"
            status_lulus = False

        # --- Perulangan FOR ---
        for pelajaran in mata_pelajaran:
            if nilai < 85:
                rekomendasi.append(tips_belajar[pelajaran])

        # --- Perulangan WHILE ---
        target = 100
        selisih = target - nilai
        langkah = 0
        while selisih > 0:
            langkah += 1
            selisih -= 5  # misalnya tiap minggu naik 5 poin

        # --- OUTPUT ---
        print("=" * 50)
        print(f"Nama: {nama}")
        print(f"Nilai: {nilai}")
        print(f"Status Lulus: {status_lulus}")
        print(f"Pesan: {pesan}")

        print("\nRekomendasi Belajar:")
        if rekomendasi:
            for r in rekomendasi:
                print(f"- {r}")
        else:
            print("- Tidak perlu rekomendasi tambahan üëç")

        print("\nMotivasi Tambahan:")
        for t in tambahan:
            print(f"- {t}")

        print(f"\nPerkiraan waktu untuk nilai sempurna: {langkah} minggu\n")


# --- Daftar siswa (list of dictionary) ---
daftar_siswa = [
    {"nama": "Vina Aprilia", "nilai": 78},
    {"nama": "Anas mutiara dewa", "nilai": 92},
    {"nama": "Rina cintya ayu", "nilai": 65},
    {"nama": "Sri Melati", "nilai": 84},
    {"nama": "Andini Putri", "nilai": 99}
]

# --- Pemanggilan Fungsi ---
sistem_penilaian_semua(daftar_siswa)
```
